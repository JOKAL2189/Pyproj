### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# using seaborn for plot

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
"""
newdf=pd.read_sql_query(QUERY,conn)
sns.scatterplot(data=newdf, x='WorkWeekHrs', y='CodeRevHrs', size='Age', hue='Age', alpha=0.5, sizes=(10, 500)) 

plt.title('WorkWeekHrs and CodeRevHrs By Age', size=12) 
plt.xlabel('WorkWeekHrs', size=8) 
plt.ylabel('CodeRevHrs', size=8)

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """ 
SELECT DatabaseDesireNextYear, COUNT(*) as count 
from DatabaseDesireNextYear 
group by DatabaseDesireNextYear 
order by count(DatabaseDesireNextYear) DESC LIMIT 5 
""" 

dfnew1=pd.read_sql_query(QUERY,conn)  
dfnew1.set_index('DatabaseDesireNextYear', inplace=True) 

colors_list=['cyan', 'yellowgreen', 'magenta', 'lightskyblue', 'lightgreen']

dfnew1['count'].plot(kind='pie', figsize=(18,8), autopct='%1.1f%%', labels=None, startangle=90, colors=colors_list, shadow=False, pctdistance=1.08)  

plt.legend(labels=dfnew1.index, loc='upper right')
plt.title('Top 5 Databases Respondents Wish To Learn') 
plt.axis('equal')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = "SELECT WorkWeekHrs, CodeRevHrs,Age from master WHERE Age BETWEEN 30 AND 35";

newdf1 = pd.read_sql_query(QUERY,conn)
newdf2 = newdf1.groupby('Age').median()

newdf2.plot(kind='bar', figsize=(12, 8), stacked=True)

plt.title('Stacked Bar Chart of The Median WorkWeekHrs and CodeRevHrs for those between Age 30 and 35')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = "SELECT ConvertedComp,Age from master WHERE Age BETWEEN 45 AND 60";

newdf3 = pd.read_sql_query(QUERY,conn)
newdf4 = newdf3.groupby('Age').median()

newdf4.plot(kind='line', figsize=(20, 6))

plt.title('Median ConvertedComp for those Between Ages 45 to 60')
plt.ylabel('ConvertedComp USD')
plt.xlabel('Age')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY = "SELECT MainBranch, COUNT(*) as MainBranchCount from master group by MainBranch";

newdf6 = pd.read_sql_query(QUERY,conn)
newdf6.plot(kind='barh', figsize=(8, 4), color='b')
plt.ylabel('MainBranch')
plt.show()

print("This is the MainBranch Count Summary:","\n","\n", newdf6)
```
